```python
import xml.etree.ElementTree as et # import etree
import numpy as np
import matplotlib.pylab as plt
%matplotlib notebook

bx = []
by = []
bz = []

p1x = []
p1y = []
p1z = []


p2x = []
p2y = []
p2z = []


p3x = []
p3y = []
p3z = []


p4x = []
p4y = []
p4z = []


p5x = []
p5y = []
p5z = []


p6x = []
p6y = []
p6z = []


p7x = []
p7y = []
p7z = []


p8x = []
p8y = []
p8z = []


p9x = []
p9y = []
p9z = []


p10x = []
p10y = []
p10z = []


def position (filename):
    tree = et.parse(filename)
    root = tree.getroot()
    
    for a in root.iter('sequences'):
        r = (float)(a.attrib['period'])
        
        #r = get(period)
        for b in root.iter('moment'):
        
        
            temp = b.attrib['locations']
            temp = temp.split(";")
            clock = (float)(b.attrib['game-clock'])
            if r == 1:
                clock = abs(clock - 720.0)
            elif r == 2:
                clock = abs(clock - 1440.0)
            elif r == 3:
                clock = abs(clock - 2160.0)
            elif r == 4:
                clock = abs(clock - 2880.0)
    
    
            for i in range(0,len(temp)):
                t = temp[i].split(",")
#       
################ BALL #####################
                if i%11 == 0:
                    bx.append((float)(t[2]))
                    by.append((float)(t[3]))
                    bz.append((float)(t[4]))
                
################# MAVS ####################                
                elif i%11 == 1:
                    
                    p1x.append((float)(t[2]))
                    p1y.append((float)(t[3]))
                    p1z.append((float)(t[4]))
                    
                elif i%11 == 2:
                    
                    p2x.append((float)(t[2]))
                    p2y.append((float)(t[3]))
                    p2z.append((float)(t[4]))
                    
                elif i%11 == 3:
                    
                    p3x.append((float)(t[2]))
                    p3y.append((float)(t[3]))
                    p3z.append((float)(t[4]))
                
                elif i%11 == 4:
                    
                    p4x.append((float)(t[2]))
                    p4y.append((float)(t[3]))
                    p4z.append((float)(t[4]))
                    
                elif i%11 == 5:
                    
                    p5x.append((float)(t[2]))
                    p5y.append((float)(t[3]))
                    p5z.append((float)(t[4]))
                    
                elif i%11 == 6:
                    
                    p6x.append((float)(t[2]))
                    p6y.append((float)(t[3]))
                    p6z.append((float)(t[4]))
                    
                elif i%11 == 7:
                    
                    p7x.append((float)(t[2]))
                    p7y.append((float)(t[3]))
                    p7z.append((float)(t[4]))
                    
                elif i%11 == 8:
                    
                    p8x.append((float)(t[2]))
                    p8y.append((float)(t[3]))
                    p8z.append((float)(t[4]))
                    
                elif i%11 == 9:
                    
                    p9x.append((float)(t[2]))
                    p9y.append((float)(t[3]))
                    p9z.append((float)(t[4]))
                    
                elif i%11 == 10:
                    
                    p10x.append((float)(t[2]))
                    p10y.append((float)(t[3]))
                    p10z.append((float)(t[4]))
                    
position('quarter1.XML')
```
