```python
####### Velocity #######
def distance_list(x_2,x_1,y_2,y_1,z_2,z_1):
    dist = 0
    dist += np.sqrt((x_2-x_1)**2 + (y_2-y_1)**2 + (z_2-z_1)**2)
    return dist
```

```python
def distance(x,y,z):
    d = []
    for i in range(1,len(x)):
        a = (distance_list(x[i],x[i-1],y[i],y[i-1],z[i],z[i-1]))
        d.append(a)
    return d
```

```python
#### 1 foot = 0.3048 meters ####
#### Want velocity in m/s in order to eventually find kinectic energy ####
#### Velocity is off because when people check in, their positions change drastically ####



def velocity(d):
    v = [] 
    for i in range(0,len(d)):
        v.append(d[i]*0.3048/0.04)
        
    return v
        
```
