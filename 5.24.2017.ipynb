```python
###################### DISTANCE FORMULA ###########################
def distance(x1, x2, y1, y2, z1, z2):
    dis = 0
    dis += np.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2-z1)**2) * 0.3048  ### Converts from ft/s ---> m/s ###
    return dis
```

```python
##################### TAKES IN 3 LISTS FOR INPUT ##################
def player_distance(x, y, z):
    d = []
    for i in range(0, len(x)):
        d.append(distance(x[i], x[i-1], y[i], y[i-1], z[i], z[i-1]))
    return d
###################################################################
print len(p3234x)
print len(p3234t)
```

```python
### percent_over_0.5 = 0



def percentage(velocity):
    
    ###################
    
    on_bench = 0
    percent_over_0_5 = 0
    percent_over_1_0 = 0
    percent_over_1_0 = 0
    percent_over_1_5 = 0 
    percent_over_2_0 = 0
    percent_over_2_5 = 0
    percent_over_3_0 = 0
    percent_over_3_5 = 0
    percent_over_4_0 = 0
    ##############################
    for i in range(0,len(velocity)):
        
        if velocity[i]>=4.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            percent_over_3_0 += 1
            percent_over_3_5 += 1
            percent_over_4_0 += 1
        
        elif velocity[i]>= 3.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            percent_over_3_0 += 1
            percent_over_3_5 += 1
            percent_over_4_0 += 1
            percent_over_3_5 += 1
            
        
        elif velocity[i]>= 3.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1            
            percent_over_3_0 += 1
        
        
        
        elif velocity[i]>= 2.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            
            
        elif velocity[i]>= 2.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            
            
        
        elif velocity[i]>= 1.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1 
            

        elif velocity[i]>= 1.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            
            
        elif velocity[i]>= 0.5:
            percent_over_0_5 += 1
            
        elif velocity[i]<= 0.1:
            on_bench += 1
            
        ########################################    
    x = len(velocity) - on_bench
    
    zero_five = 1.0*(float)(percent_over_0_5)/x
    
    one = 1.0*(float)(percent_over_1_0)/x
    one_five = 1.0*(float)(percent_over_1_5)/x
    
    two = 1.0*(float)(percent_over_2_0)/x
    two_five = 1.0*(float)(percent_over_2_5)/x
    
    three = 1.0*(float)(percent_over_3_0)/x
    three_five = 1.0*(float)(percent_over_3_5)/x
    
    four = 1.0*(float)(percent_over_4_0)/(float)(x)

    return zero_five, one, one_five, two, two_five, three, three_five, four
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
```

```python
#########
# Average Velocity #
#########



def avg_vel(velocity):
    total_speed = 0 
    seconds_on_court = 0
    
    for i in range(0,len(velocity)):
        if velocity[i] > 0:
            total_speed += velocity[i]
            seconds_on_court += 1
    
    return (1.0*total_speed/seconds_on_court)
```

```python
def seconds_on_court(velocity):
    
    ###################
    
    on_bench = 0
    percent_over_0_5 = 0
    percent_over_1_0 = 0
    percent_over_1_0 = 0
    percent_over_1_5 = 0 
    percent_over_2_0 = 0
    percent_over_2_5 = 0
    percent_over_3_0 = 0
    percent_over_3_5 = 0
    percent_over_4_0 = 0
    ##############################
    for i in range(0,len(velocity)):
        
        if velocity[i]>=4.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            percent_over_3_0 += 1
            percent_over_3_5 += 1
            percent_over_4_0 += 1
        
        elif velocity[i]>= 3.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            percent_over_3_0 += 1
            percent_over_3_5 += 1
            percent_over_4_0 += 1
            percent_over_3_5 += 1
            
        
        elif velocity[i]>= 3.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1            
            percent_over_3_0 += 1
        
        
        
        elif velocity[i]>= 2.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            percent_over_2_5 += 1
            
            
        elif velocity[i]>= 2.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1
            percent_over_2_0 += 1
            
            
        
        elif velocity[i]>= 1.5:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            percent_over_1_5 += 1 
            

        elif velocity[i]>= 1.0:
            
            percent_over_0_5 += 1
            percent_over_1_0 += 1
            
            
        elif velocity[i]>= 0.5:
            percent_over_0_5 += 1
            
        elif velocity[i]<= 0.1:
            on_bench += 1
            
        ########################################    
    x = 25.00 * 60.00 #25 captures per second, 60 seconds per minute
    
    zero_five = 1.0*(float)(percent_over_0_5)/x
    
    one = 1.0*(float)(percent_over_1_0)/x
    one_five = 1.0*(float)(percent_over_1_5)/x
    
    two = 1.0*(float)(percent_over_2_0)/x
    two_five = 1.0*(float)(percent_over_2_5)/x
    
    three = 1.0*(float)(percent_over_3_0)/x
    three_five = 1.0*(float)(percent_over_3_5)/x
    
    four = 1.0*(float)(percent_over_4_0)/(float)(x)

    return zero_five, one, one_five, two, two_five, three, three_five, four
            
            
```

```python
def k_energy(velocity,mass): # when entering mass and velocity, make sure player numbers match up!
    #k_instantaneous = []
    #for i in range(0,len(velocity)):
     #   k_instantaneous.append( 0.5 * mass * velocity[i]**2 )
    
    #return k_instantaneous*1.0/len(velocity)
    
    
    counter = []
    
    
    for i in range(0,len(velocity)):
        if velocity[i]>=0:
            
            counter.append(0.5*mass*velocity[i]*velocity[i])
        
    return sum(counter)
```
